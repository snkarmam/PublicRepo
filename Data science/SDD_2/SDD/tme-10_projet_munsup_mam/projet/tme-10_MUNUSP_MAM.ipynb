## Présentation des données 
Le jeu de données est composé de 16 attributs et ils sont tous entiers. Ces points qui se connectent pour réprésenté les chiffres 
variés de 0 à 9. Chaque donnée de X représente alors les 16 points qui permet de dessiner les chiffre , et Y représente la classe qui 
permet d'indiquer le chifrre de chaque données de X

### Extraction des données
Au début on a essayé à utiliser les données directement dans les classifieur, mais il s'est posé beaucoup des problème avec les dimensions 
des données X et Y. On a alors convertir les données en array pour faciliter la manipulation 

```python
nb_lig,nb_col = X.shape
X_np = np.array(X)
Y_np = np.array([y["Class"][i] for i in range(nb_lig)])
Y_np
```

### Visualisation des données
Pour mieux comprendre et visualiser chaque données, on a utilisé la fonctionnement de pyplot ci-dessous pour voir si on trace les 16 points 
il représente bien le chiffre comme indiqué dans les labels de Y.
Par exemple la premiere données de X, selon le label de Y, il représente le numéro 8

```python
import matplotlib.pyplot as plt
for i in range(0,13,2):
    x0 = X.iloc[0][i]
    y0=X.iloc[0][i+1]
    x1 = X.iloc[0][i+2]
    y1 = X.iloc[0][i+3]
    plt.plot([x0,x1],[y0,y1])
plt.show()

```

```python
def tracer_accuracy(accuracies, title="Courbe de Accuracy", xlabel="Iteration", ylabel="Accuracy"):

    plt.figure(figsize=(8, 5))
    plt.plot(accuracies, marker='o', linestyle='-', color='b')

    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)

    plt.grid(True)

    plt.xticks(range(len(accuracies)), [str(i+1) for i in range(len(accuracies))])

    max_acc = max(accuracies)
    max_index = accuracies.index(max_acc)
    plt.scatter(max_index, max_acc, color='red')  
    plt.annotate(f'Max: {max_acc:.2f}', (max_index, max_acc), textcoords="offset points", xytext=(0,10), ha='center')

    # Show the plot
    plt.show()
```

## premier expérimenttation avec le classifieur KNN

On a alors appliqué les données X et Y avec le classifieurKNN (Ici on utilise 5 plus proche voisins) 

```python

np.random.seed(42)   # on prend 42 comme graine
k=5
knn = classif.ClassifierKNN(16,5)
perf, taux_moyen, taux_ecart = ev.validation_croisee(knn,(X_np,Y_np),10) 

# Classification du dataset (contrôle de predict et score):
print("Classification KNN des exemples du dataset:")
print(f'Analyse perf: moyenne: {taux_moyen:0.4f}\tecart: {taux_ecart:0.4f}')

tracer_accuracy(perf, "Accuracy pour le class KNN")
```

## le classifieur KNN multi-class
Il a atteint une précision très élevée de 99,5997% sur le jeu de données, 
indiquant une performance excellente dans la distinction entre différentes classes. 

```python
# Normalisation des données
X_norm = clust.normalisation(X)
X_norm

y = np.array(y).flatten()
X_norm = np.array(X_norm)
y

permut = np.random.permutation(np.arange(y.size))
X_permut = X_norm[permut]
Y_permut = y[permut]
Y_permut
```

```python
import copy
n = 10
knn = classif.ClassifierKNN_MC(X.shape[1], 5, 11)
Acc_train = []
Acc_test = []
for i in range(n):
    X_train, Y_train, X_test, Y_test = ev.crossval(X_permut, Y_permut, n, i)
    trained = copy.deepcopy(knn)
    trained.train(X_train, Y_train)
    Acc_train.append(trained.accuracy(X_train, Y_train))
    Acc_test.append(trained.accuracy(X_test, Y_test))
```

```python
print('Performances pour les trains:\n', Acc_train)
print('moyenne et variance:\n', ev.analyse_perfs(Acc_train))

tracer_accuracy(Acc_train, "Accuracy des exemples train de la classe KNN_MC")
```

```python
print('Performances pour les tests : \n', Acc_test)
print('moyenne et variance : \n', ev.analyse_perfs(Acc_test))

tracer_accuracy(Acc_test, "Accuracy des exemples test de la class KNN_MC")
```

## le classifieur perceptronBiaiset Multi OAA


```python
import copy
perceptronMulti = classif.ClassifierMultiOAA(classif.ClassifierPerceptronBiais(X.shape[1], 1e-3, True))

Acc_train = []
Acc_test = []
n = 5

for i in range(n):
    X_train, Y_train, X_test, Y_test = ev.crossval(X_permut, Y_permut, n, i)
    trained = copy.deepcopy(perceptronMulti)
    perceptronMulti.train(X_train, Y_train)
    Acc_train.append(perceptronMulti.accuracy(X_train, Y_train))
    Acc_test.append(perceptronMulti.accuracy(X_test, Y_test))


print('Performances en apprentissage :', Acc_train)
print('Moyenne et variance :', ev.analyse_perfs(Acc_train))
tracer_accuracy(Acc_train, "Accuracy des exemples train de la classe KNN_MC")
print('Performances en test :', Acc_test)
print('Moyenne et variance :', ev.analyse_perfs(Acc_test))
tracer_accuracy(Acc_test, "Accuracy des exemples test de la class KNN_MC")
```

## arbre de Décision (ClassifierArbreNumerique) 

 Un classifieur arbre de décision a été construit et entraîné sur le jeu de données.Ce modèle a démontré une haute performance avec une précision
moyenne de 95,2207% à travers une validation croisée en 10 pas.

## arbre de Décision (ClassifierArbreDecision) 

Pour l'expérimentation avec le classifieur arbre de décision où n=5, nous observons un résultat très contrasté entre les performances en apprentissage et celles en test.
Performances en apprentissage
    Les résultats obtenus pour l'ensemble d'apprentissage sont parfaits avec une précision de 100%. Ceci est attendu pour un arbre de décision qui peut facilement mémoriser les données d'entraînement.
Performances en test
    Cependant, la performance sur l'ensemble de test est significativement plus basse avec une précision d'environ 35.50%. Cela suggère un surajustement sévère du modèle aux données d'entraînement, entraînant une faible capacité de généralisation sur de nouvelles données.


```python
n = 5
Acc_train = []
Acc_test = []
for i in range(n):
    X_train, Y_train, X_test, Y_test = ev.crossval(X_permut, Y_permut, n, i)
    arbre = classif.ClassifierArbreDecision(X.shape[1], .0, X.columns)
    trained = copy.deepcopy(arbre)
    trained.train(X_train, Y_train)
    Acc_train.append(trained.accuracy(X_train, Y_train))
    Acc_test.append(trained.accuracy(X_test, Y_test))
    break

print('Performances en apprentissage :\n', Acc_train)
print('Moyenne et variance :\n', ev.analyse_perfs(Acc_train))


print('\n')
print('Performances en test :\n', Acc_test)
print('Moyenne et variance :\n', ev.analyse_perfs(Acc_test))

# Initialiation d'un arbre pour le dataset X:
arbre = classif.ClassifierArbreDecision(len(x_noms), 0.0, x_noms)

# Construction de l'arbre de décision à partir du dataset X
arbre.train(X_permut,Y_permut)

# Construction de la représentation graphique (affichage)
graphe_arbre_chiffre = gv.Digraph(format='png')
arbre.affiche(graphe_arbre_chiffre)
graphe_arbre_chiffre
```
